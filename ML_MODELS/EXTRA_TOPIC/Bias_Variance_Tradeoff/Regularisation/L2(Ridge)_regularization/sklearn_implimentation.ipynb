```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
# loading a dataset 
from sklearn.datasets import load_diabetes
X_train, Y_train = load_diabetes(as_frame= True, return_X_y=True)
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X_train, Y_train, train_size=0.8, random_state=42)
```

```python
#Formula
"""
y = mX + c
m = np.mean((xi - np.mean(x))*(yi - np.mean(yi))/((xi - np.mean(x))**2) + alpha)
c = np.mean(x) - m*np.mean(y)
"""
```

```python
rdg = Ridge()
```

```python
rdg.fit(X_train, Y_train)
```

```python
print(rdg.coef_, rdg.intercept_)
```

```python
# Intercept : 152.133484162896 
# Coeff     : [  29.46611189  -83.15427636  306.35268015  201.62773437    5.90961437
#   -29.51549508 -152.04028006  117.3117316   262.94429001  111.87895644]
li = [0 , 0.1 , 0.5, 0.9, 1, 1.1 , 1.5, 1.9, 2 ]
for i in li:
    print(i)
    rdg = Ridge(alpha=i)
    rdg.fit(X_train, Y_train)
    y_pred = rdg.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = root_mean_squared_error(y_test, y_pred)
    print(f"{rdg.intercept_}\n {rdg.coef_}\nMAE: {mae}\nMSE: {mse}\nRMSE: {rmse}\nr2: {r2}\n\n")
```
