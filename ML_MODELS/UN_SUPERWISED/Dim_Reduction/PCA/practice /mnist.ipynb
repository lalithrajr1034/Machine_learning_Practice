```python
df = pd.read_csv("mnist_test.csv")
```

```python
Y = df.iloc[:,0]
X = df.iloc[:,1:]
```

```python

# import the file
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
def best_pc_taker():
    for i in range(1,df.shape[1]):
        columns = PCA(n_components=i)
        prin_comp = columns.fit_transform(X)

        #Model building
        model = GaussianNB()
        x_train, x_test, y_train, y_test = train_test_split(prin_comp, Y, train_size=0.2, random_state=42)
        model.fit(x_train, y_train)
        test = model.predict(x_test)
        print(f"col_no:{i} --  {np.round(accuracy_score(test, y_test)*100,2)}%")
```

```python
best_pc_taker()
```

<h1>To find a Eigen value and Eigen vector</h1>

```python
columns = PCA(n_components=None)
columns.fit_transform(X)
```

```python
eig_value = columns.explained_variance_
```

```python
eig_value_perc = columns.explained_variance_ratio_
```

```python
for i,j in enumerate(np.cumsum(eig_value_perc)):
    if j>=0.90:
        print("principal component's:",i+1)
        break
```
