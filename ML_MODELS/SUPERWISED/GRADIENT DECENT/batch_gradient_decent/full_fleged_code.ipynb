```python
class GradentDecent:
    def __init__(self, epoches = 300, lr = 0.5):
        self.epoches = epoches
        self.lr = lr
        self.coef_ = None
        self.intercept_ = 0
        self.coef_der = None

    def fit(self, X_train, Y_train):
        Y_train = Y_train.values.flatten()
        self.coef_ = np.ones(X_train.shape[1])  
        le = len(X_train)
        print(self.coef_)
        for i in range(self.epoches):
            y_hack = np.dot(self.coef_ , X_train.T) + self.intercept_
            intercept_der = -2/le * np.mean(Y_train - y_hack)
            self.intercept_ = self.intercept_ - self.lr * intercept_der

            deravative = -2/le* np.dot((Y_train - y_hack), X_train)
            self.coef_ = self.coef_ - self.lr* deravative
            print(self.intercept_, self.coef_)
```
