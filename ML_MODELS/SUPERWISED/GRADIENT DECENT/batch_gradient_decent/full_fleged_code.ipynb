```python
class GradentDecent:
    def __init__(self, epoches = 1, lr = 0.5):
        self.epoches = epoches
        self.lr = lr
        self.coef_ = None
        self.intercept_ = 0
        self.coef_der = None
        self.a = None
    def fit(self, X_train, Y_train):
        self.coef_ = np.ones(X_train.shape[1])  
        le = len(X_train)
        for i in range(self.epoches):
            y_hack = np.dot(self.coef_ , X_train.T) + self.intercept_
            intercept_der = -2/le * np.mean(Y_train.values - y_hack)
            self.intercept_ = self.intercept_ - self.lr * intercept_der

            deravative = -2/le* np.dot((Y_train.values - y_hack), X_train)
            self.coef_ = self.coef_ - self.lr* deravative
            self.a = y_hack
```

```python
gd = GradentDecent()
gd.fit(X,y)
li = []
for i in range(len(gd.a)):
    li.append(float((y.values[i] - gd.a[i])[0])
li
```
