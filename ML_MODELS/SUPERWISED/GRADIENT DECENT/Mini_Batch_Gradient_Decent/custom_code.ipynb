```python
import numpy as np
import random

class MBG:
    def __init__(self, group=10, epoches=100, lr=0.01):
        self.group = group
        self.epoches = epoches
        self.lr = lr
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X_train, Y_train):
        # Convert to NumPy and fix shapes
        X = np.array(X_train)
        y = np.array(Y_train).reshape(-1)  # force 1D

        n_samples, n_features = X.shape
        batch_size = max(1, n_samples // self.group)

        # Initialize parameters
        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0.0

        for epoch in range(self.epoches):
            print(self.intercept_ , self.coef_)
            for _ in range(self.group):
                ind = random.sample(range(n_samples), batch_size)

                X_b = X[ind]
                y_b = y[ind]

                # Prediction
                y_hat = X_b @ self.coef_ + self.intercept_

                # Gradients
                error = y_b - y_hat
                d_intercept = -2 * np.mean(error)
                d_coef = -2 * (X_b.T @ error) / batch_size

                # Update
                self.intercept_ -= self.lr * d_intercept
                self.coef_ -= self.lr * d_coef

        print("Intercept:", self.intercept_)
        print("Coefficients:", self.coef_)

    def predict(self, X_test):
        X_test = np.array(X_test)
        return X_test @ self.coef_ + self.intercept_
```
