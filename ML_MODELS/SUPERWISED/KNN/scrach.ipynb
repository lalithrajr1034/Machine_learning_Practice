```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris(return_X_y=True)
```

```python
df1 = pd.DataFrame(df[0])
df2 = pd.DataFrame(df[1])
```

```python
df1
```

```python
li = []
for i in range(len(df1)):
    lis = []
    for j in range(len(df1)):
        if i != j:
            lis.append(np.sqrt(np.sum(df1.iloc[i] - df1.iloc[j])**2))
    li.append(lis)
```

```python
min_values = []
no_of_val = 5
for i in range(len(df1)):
    min_values.append(np.argsort(li[i])[:no_of_val])
```

```python
min_values
```

```python
def predict(values):
    pred_li = []
    for i in range(len(values)):
        pred_li.append(np.max(df2.iloc[values[i]]))
    return pred_li
```

```python
predict(min_values)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(df2, predict(min_values))
```
