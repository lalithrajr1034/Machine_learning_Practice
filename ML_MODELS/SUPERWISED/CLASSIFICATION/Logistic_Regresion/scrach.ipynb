```python
def sigmoid(value):
    return 1/(1+np.exp(-value))
```

```python
class LogisticRegresion:
    def __init__(self, lr = 0.5):
        self.lr = lr
        self.coef_ = None
        self.inter_= None
    def fit(self, x_train, y_train):
        x_train = np.insert(x_train, 0, 1, axis = 1)
        coef = np.ones(x_train.shape[1])
        
        for i in range(100):
            # here we are calculatin a sigmoid function that is y_hack
            """print(coef.shape,x_train.shape)"""
            y_hak = sigmoid(x_train @ coef)

            # Wnew = Wold - lr*(dl/dw)
            a = np.array(list(y_train.iloc[:,0])) - np.array(y_hak)

            coef = coef + (self.lr*np.dot(a,x_train)/x_train.shape[0])
        self.coef_ = coef[1:]    
        self.inter_= coef[0]
        return self.inter_, self.coef_
        
    def predect(self, x_test):
        value = sigmoid((x_test @ self.coef_) + self.inter_)
        return np.where(value>=0.5, 1, 0)
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42) 
lg = LogisticRegresion()
lg.fit(X_train, Y_train)
```

```python
y_pred = lg.predect(x_test)
```
