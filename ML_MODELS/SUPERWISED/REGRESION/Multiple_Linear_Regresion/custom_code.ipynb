```python
X = df.iloc[:,:-1]
Y = df.iloc[:, -1]
```

```python
class MultipleLiner:
    def __init__(self):
        self.coef = None
        self.pred = None
    def fit(self, x_train, y_train):
        x_train = np.insert(x_train, 0, 1, axis = 1)
        self.coef = np.dot(np.dot(np.linalg.inv(np.dot(x_train.T, x_train)),x_train.T),y_train)
        return self.coef
    def predect(self, x_test):
        x_test = np.insert(x_test, 0, 1, axis = 1)
        self.pred = np.dot(x_test, self.coef)
        return self.pred
```

```python
a = MultipleLiner()
```

```python
a.fit(X, Y)
a.coef[0]
```
