```python
train, test = df.iloc[:,0], df.iloc[:,1]
```

```python
class Simple_LG:
    def __init__(self):
        self.m  = None
        self.c =   None
        
    def fit(self, train, test): 
        # Slope
        mean_train, mean_test = train.mean(), test.mean()
        num, den = 0, 0
        for i in range(len(train)):
            num = num + (train[i]-mean_train)*(test[i]-mean_test)
            den = den + (train[i]-mean_train)**2
        self.m = num/den

        # Intercept
        self.c = mean_test - self.m*mean_train        
    def predect(self, train):
        return (self.m * train) + self.c
```

```python
model = Simple_LG()
```

```python
model.fit(train, test)
```
