```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.metrics import confusion_matrix, accuracy_score,r2_score
```

```python
x,y = load_diabetes(as_frame= True, return_X_y= True)
```

```python
x
```

```python
y
```

```python
X_train, x_test, Y_train, y_test = train_test_split(x,y, random_state=42, train_size=0.8)
```

```python
model1 = DecisionTreeRegressor(max_depth=6, min_samples_split=10)
model2 = DecisionTreeRegressor(max_depth=5, min_samples_split=15)
model3 = DecisionTreeRegressor(max_depth=3, min_samples_split=20,max_features=3)
model4 = DecisionTreeRegressor(max_depth=4, min_samples_split=20)
model5 = DecisionTreeRegressor(max_depth=8, min_samples_split=11)
```

```python
def sample_value(data, samp = 250):
    return data.sample(samp).index
```

```python
a = sample_value(X_train)
model1.fit(X_train.loc[a], Y_train.loc[a])

a = sample_value(X_train)
model2.fit(X_train.loc[a], Y_train.loc[a])

a = sample_value(X_train)
model3.fit(X_train.loc[a], Y_train.loc[a])

a = sample_value(X_train)
model4.fit(X_train.loc[a], Y_train.loc[a])

a = sample_value(X_train)
model5.fit(X_train.loc[a], Y_train.loc[a])
```

<h3> Results on train data</h3>

```python
# Model 1
r2_score(Y_train,model1.predict(X_train))
```
