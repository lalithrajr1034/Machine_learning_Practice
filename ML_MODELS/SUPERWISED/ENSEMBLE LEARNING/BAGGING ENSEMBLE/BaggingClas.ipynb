```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
```

```python
X,y = load_breast_cancer(as_frame=True, return_X_y= True)
```

```python
X
```

```python
y
```

```python
model = BaggingClassifier(
    estimator= DecisionTreeClassifier(),
    n_estimators= 10,
    max_samples=0.7,
    bootstrap=True, # with replacement
    oob_score= True,
    n_jobs= -1,
    random_state= 42,
    verbose= 1
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, Y_train , y_test = train_test_split(X, y, random_state=42, train_size=0.8)
```

```python
X_train
```

```python
x_test
```

```python
model.fit(X_train, Y_train)
```

```python
train = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_train, train)
```

```python
accuracy_score(y_test, model.predict(x_test))
```

```python
svm_model = BaggingClassifier(
    estimator= SVC(),
    n_estimators=10,
    max_samples=0.7,
    bootstrap= True,
    oob_score= False,
    n_jobs= -1,
    random_state= 42,
    verbose= 1
)
```

```python
svm_model.fit(X_train,Y_train)
```

```python
accuracy_score(Y_train, svm_model.predict(X_train))
```

```python
accuracy_score(y_test, svm_model.predict(x_test))
```

<h1>Bootstraping = False<h1/>

```python
fa_model = BaggingClassifier(
    estimator= SVC(),
    n_estimators= 20,
    max_samples= 0.75,
    bootstrap= False,
    n_jobs= -1,
    random_state= 42,
    verbose= 1
)
```

```python
fa_model.fit(X_train, Y_train)
```

```python
accuracy_score(y_test, fa_model.predict(x_test))
```

```python
accuracy_score(Y_train, fa_model.predict(X_train))
```
