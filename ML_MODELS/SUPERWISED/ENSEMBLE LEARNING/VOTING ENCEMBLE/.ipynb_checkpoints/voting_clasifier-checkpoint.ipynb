```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
data1, data2 = load_breast_cancer(as_frame=True, return_X_y=True)
```

```python
data1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, Y_train, y_test = train_test_split(data1, data2, train_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
gnb = GaussianNB()
svm = SVC(probability=True)
dtree = DecisionTreeClassifier()
lgr = LogisticRegression(max_iter=1000)
```

```python
a = [('gnb', gnb), ('svm', svm), ('dtree', dtree), ('lgr', lgr)]
```

```python
for i in a:
    print(cross_val_score(i[1], data1, data2, scoring='accuracy', cv = 5).mean())
```

```python
for i in a:
    model = i[1].fit(X_train, Y_train)
    pred = model.predict(x_test)
    print(f'{str(i[1])[:-2]},    {accuracy_score(y_test, pred)}')
```

<h1> Hard Voting</h1>

```python
stf = StratifiedKFold(n_splits=5)
```

```python
vot = VotingClassifier(estimators = a, voting = 'hard')
```

```python
result = cross_val_score(vot , data1, data2,cv = stf, scoring= 'accuracy')
```

```python
result
```

```python
result.mean() # this is the result of a voting clasifier
```

<h1>soft voting</h1>

```python
a
```

```python
for i in a:
    print(cross_val_score(i[1], data1, data2, cv=5, scoring='accuracy').mean())
```

```python
voting = VotingClassifier(a, voting= 'soft')
```

```python
values = cross_val_score(voting, data1, data2, cv=stf)
```

```python
values.mean()
```

<h1>Weighth<h1/>

```python
model = [('gnb', gnb), ('svm', svm), ('dtree', dtree), ('lgr', lgr)]
li = []
a = 5
for i in range(1, a):
    for j in range(1, a):
        for k in range(1, a):
            for l in range(1, a):
                val = VotingClassifier(model, voting = 'soft', weights=[i, j, k, l])
                b = cross_val_score(val, data1, data2, cv=5, scoring='accuracy')
                li.append(f'{i}, {j}, {k}, {l}:{b}')
```
