```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Sample data
df = pd.DataFrame({
    'Age': [25, np.nan, 30, 22, np.nan],
    'Income': [50000, 60000, 55000, np.nan, 52000],
    'Target': [1, 0, 1, 0, 1]
})

X = df[['Age', 'Income']]
y = df['Target']

# Pipeline
pipe = Pipeline([
    ('imputer', SimpleImputer()),  # we will tune this
    ('model', RandomForestClassifier())
])

# Grid: try different strategies for missing values
param_grid = {
    'imputer__strategy': ['mean', 'median', 'most_frequent', 'constant'],
    'imputer__fill_value': [-10],  # only used if strategy='constant'
    'model__n_estimators': [50, 100]
}

# GridSearchCV
grid = GridSearchCV(pipe, param_grid, cv=3)
grid.fit(X, y)

print("Best strategy:", grid.best_params_)
```
