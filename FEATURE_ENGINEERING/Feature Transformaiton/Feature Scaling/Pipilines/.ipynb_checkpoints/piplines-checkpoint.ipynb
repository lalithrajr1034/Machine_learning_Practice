```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day29-sklearn-pipelines/train.csv')
```

```python
df.head(1)
```

```python
df.columns
```

```python
df = df[['Survived', 'Pclass', 'Sex', 'Age','Fare', 'Embarked']]
df.head(2)
```

```python
df['Embarked'].value_counts()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
col1 = ColumnTransformer(
    [('age', SimpleImputer(strategy='mean'),[3]),
     ('embarked', SimpleImputer(strategy="most_frequent"), [5])
    ],remainder='passthrough'
)
```

```python
#Nominal encoding
 
col2 = ColumnTransformer(
    [
        ('onehot_nummer', OneHotEncoder(), [0,2]),
        ('onehot_catego', OneHotEncoder(handle_unknown='ignore', categories=[['S','C','Q']]), [5])
    ],remainder = 'passthrough'
)
# 2+3+3 = 8 , 5 columns are extra
```

```python
# instead of 1 we use 2 because of increasing number of column
col3 = ColumnTransformer(
    [('pclass', OrdinalEncoder(), [2])],
    remainder = 'passthrough'
)
```

```python
#8 col
col4 = ColumnTransformer(
    [('standard scaling',StandardScaler(), [6,7])]
)
```

<h13>Pipeline</h3>

```python
pipe = Pipeline([
    ('missing',col1),
    ('nominal',col2),
    ('ordinal',col3),
    ('scaling',col4)
])
```

```python
pipe.fit_transform(df)
# this was a hecktick and it is not scalable in long run so make a encoding column in last and make a proper plan
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# before ['Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'Embarked']
colt1 = ColumnTransformer(transformers=[
    ('age_simpleimp', SimpleImputer(),[3]),
    ('embark_simple', SimpleImputer(strategy="most_frequent"),[5])
],remainder='passthrough')

```

```python
# before ['Age', 'Embarked', 'Survived', 'Pclass', 'Sex', 'Fare']
colt2 = ColumnTransformer(transformers=[
    ('std_scalar',StandardScaler(),[0,5])
],remainder='passthrough')

```

```python
# before ['Age', 'Fare', 'Embarked', 'Survived', 'Pclass', 'Sex']
colt3 = ColumnTransformer(transformers=[
    ('one_hot_num', OneHotEncoder(),[3]),
    ('one_hot_emba', OneHotEncoder(categories=[['S', 'C', 'Q']], handle_unknown='ignore'),[2]),
    ('one_hot_sex', OneHotEncoder(categories=[['male','female']]),[5]),
    ('ordinal', OrdinalEncoder(categories=[[1,2,3]]),[4])
],remainder='passthrough')

```

```python
pipe = Pipeline(steps=[
    ('col1',colt1),
    ('col2',colt2),
    ('col3',colt3)
])
```

```python
# ['Survived', 'Embarked', 'Sex', 'Pclass', 'Age', 'Fare']
pipe.fit_transform(df)[1]
```
