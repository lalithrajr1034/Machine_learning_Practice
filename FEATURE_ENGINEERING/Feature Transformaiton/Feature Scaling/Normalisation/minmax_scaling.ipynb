```python
# Min max scaling 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day25-normalization/wine_data.csv", usecols= [0,1,2], header=None)
df.columns = ['Class Label', 'Alcohol', 'Malic Acid']
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
sns.kdeplot(df['Malic Acid'])
```

```python
x_train, x_test = train_test_split(df.iloc[:, 1:], train_size=0.2)
x_test,x_train 
```

```python
val = MinMaxScaler()
val.fit(x_train)

x_train_min = val.transform(x_train)
x_test_min  = val.transform(x_test)
```

```python
sns.scatterplot(x_test)
plt.plot()
```
