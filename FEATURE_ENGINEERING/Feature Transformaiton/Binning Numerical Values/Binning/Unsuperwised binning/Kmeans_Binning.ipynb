```python
df = sns.load_dataset('titanic')
```

```python
df.fare
```

```python
from sklearn.preprocessing import KBinsDiscretizer
no = 10
bin = KBinsDiscretizer(n_bins=no, encode='ordinal', strategy='kmeans')
```

```python
bi = bin.fit_transform(np.array(df.fare).reshape(len(df), 1))
```

```python
pd.DataFrame(bi).value_counts().plot(kind = 'bar')
```
