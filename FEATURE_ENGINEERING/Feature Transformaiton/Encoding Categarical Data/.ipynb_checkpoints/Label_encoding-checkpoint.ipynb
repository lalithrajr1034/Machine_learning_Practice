```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#Split the dependent column in to a train test split
x_train, x_test, y_train, y_test = train_test_split("here you have to pass the data", test_size = 0.2)
x_train, y_train

#Fit to a model 
laen = LabelEncoder() #her we need not to clasify catagory on priority
y_train = y_train.ravel()
y_test  = y_test.ravel()
laen.fit(y_train)

#Encode a catagorical column
laen.classes_
laen.transform(y_train)
laen.transform(y_test)

#We get a encoded output
y_train, y_test
```
