```python
# instead of a 2 column we can make a single column that is a bmi 
```

```python
"""
so the formula  for this is 
BMI = ((weight)*703)/(height)**2
"""
```

```python
li = []
for i in range(len(df)):
    a = (df.iloc[i,1]*703)/np.square(df.iloc[i,0]).item()
    li.append(a)
```

```python
li
```

```python
df['BMI'] = li
df = df['BMI']
```

```python
df
```

```python
df = df.round(2)
df
```
